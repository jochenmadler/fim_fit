```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
```

```python
# read in hourly capacity factors from excel
hourly_cap_factors_xlsx = pd.read_excel("PV_hourly_capacity.xlsx")
pd.options.display.float_format = '{:,.4f}'.format
hourly_factors = hourly_cap_factors_xlsx.rename(columns={"Capacity factor": "capacity_factor"})
hourly_factors
```

```python
# for each row, add three empty ones
empty_rows = 4
hourly_factors.index = range(0, empty_rows*len(hourly_factors), empty_rows)
df = hourly_factors.reindex(index=range(empty_rows*len(hourly_factors)))
# df
```

```python
# generate 15min intervall series
start_date = datetime(2019,1,1)
quarter_hourly = [start_date + timedelta(minutes=15*x) for x in range(0,35040)]
quarter_hourly_list = [x.strftime('%Y-%m-%d %H:%M') for x in quarter_hourly]
# generate dataframe
quarter_hourly_list = pd.DataFrame(quarter_hourly_list, columns=['datetime'])
#quarter_hourly_list
```

```python
# insert quarter-hourly list into df
df['datetime'] = quarter_hourly_list['datetime']
df_new = df[['datetime', 'capacity_factor']]
df_new = df_new.rename({'capacity_factor' : 'hourly_capacity_factor'}, axis='columns')
# df_new
```

```python
# interpolate hourly to quarter-hourly values: linear interpolation
# since factors represent *hourly* capacity, a quarter hour only yields 1/4 the power
df_quarterly = df_new.interpolate(method='linear')
df_quarterly = df_quarterly.assign(quarterly_capacity_factor = lambda x: (x['hourly_capacity_factor'] / 4))
df_quarterly = df_quarterly[['datetime', 'quarterly_capacity_factor']]
df_quarterly.head(10)
```

```python
# write to xlsx
df_quarterly.to_excel(r'C:\Users\joche\ProgrammingProjects\04 FIM-RC\PV_wind power simulation\PV_quarterly_capacity.xlsx', index=False, header=True)
```
