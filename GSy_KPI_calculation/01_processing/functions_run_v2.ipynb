# KPI Calculation
---
There are four KPIs calculated. As GSy's model is hierarchical, each KPI is calculated for several levels
* **Average electricity price and cost per timestep** (done)
    - germany
    - regions
    - ecs
    - assets
* **Total electricity costs** (done)
    - germany
    - market_maker
    - regions
    - ecs
    - other_assets
* **Share of renewable energy consumed** (tbd)
    - germany
    - regions
    - ecs
* **C02 emissions of energy consumed** (tbd)
    - germany
    - regions
    - ecs

#### Use case 0:
In use case 0, p2p trading is disabled. Every asset can trade with the external market through its own MM-entity. Therefore, every asset procures its energy from the German electricity mix. However, some houses have pv feed-in, reducing their energy demanded from the market and thereby decreasing their share of external, grey electricity consumed at certain hours of the day.

Trading happens along the following hierarchy: MM -> house -> hh_load.

```python
os.chdir(home_path)
os.chdir('usecase_0')
df = f.share_renewables(os.getcwd())
df
```

```python
df[['region-6-ec1-house-0_id-1-ev-commuter', 'region-6-ec1-house-0_region-6-id-81-pv-feedin', 'region-6-ec1-house-0_region-6-id-457-hh-1-household-electric-load']]
```

```python
fig, ax = plt.subplots(figsize = (20,6), nrows=3, ncols=2)
fig.suptitle('Usecase 0: Share of grey electricity (in %) per hour')
df[['region-6-ec0-house-wind']].plot(ax = ax[0][0])
df[['region-6-ec0-house-wind_region-6-id-6-wind-feedin']].plot(ax = ax[0][1])
df[['region-6-ec1-house-0']].plot(ax = ax[1][0])
df[['region-6-ec1-house-0_id-1-ev-commuter', 'region-6-ec1-house-0_region-6-id-81-pv-feedin', 'region-6-ec1-house-0_region-6-id-457-hh-1-household-electric-load']].plot(ax = ax[1][1])
df[['region-6-ec1-house-4']].plot(ax = ax[2][0])
df[['region-6-ec1-house-4_id-2-ev-non-commuter', 'region-6-ec1-house-4_region-6-id-61-heatpump-loadprofile', 'region-6-ec1-house-4_region-6-id-461-hh-1-household-electric-load']].plot(ax = ax[2][1])
for col in ax:
    for row in col:
        for i,t in enumerate(row.get_xticklabels()):
            if (i % 12) != 0:
                t.set_visible(False)
plt.legend()
plt.tight_layout()
plt.show()
```

#### Use case 4:
In use case 4, p2p trading is enabled on an energy community (ec) level. Every ec has its own MM. The trading happens along the following hierarchy: ec_mm -> ec -> house -> hh_load.

However, there is no information on p2p trading of e.g. between ecs or among houses within an ec. Therefore, ecs are either energy-producing if they contain wind (ec0) or energy-demanding otherwise. The only source of renewable energy trading is within a household, when pv production party offsets demand. In these rare cases, the share of grey electricity is < 1.

```python
os.chdir(home_path)
os.chdir('usecase4_edu')
df = f.share_renewables(os.getcwd())
df
```
