```python
fig, axes = plt.subplots(figsize = (20,12), nrows=2, ncols=2)
ctr = 1
for rows in axes:
    for cols in rows:
        sns.barplot(x = 'name', y = 'total', data=total_cost_df_list[ctr],ax = cols)
        plt.setp(cols.get_xticklabels(), rotation = 45)
        ctr += 1
plt.show()
```

```python
df_names = ["df_total", "df_germany", "df_marketm", "df_regions", "df_ecomms", "df_members"]
ctr = 0
for df_temp in total_cost_df_list:
    t = 1 if len(df_temp) >= 30 else 2
    print(f'{df_names[ctr]}:\t\t length: {len(df_temp)}', '\t' * t, f'total_cost: {df_temp.total.sum()}')
    if ctr < 1:
        print('-' * 65)
    ctr += 1
print('-' * 65)
print(
    f'df_sum: \t\t length: {np.sum([len(df) for df in total_cost_df_list[1:]])}\t total_cost: {np.sum([df.total.sum() for df in total_cost_df_list[1:]])}')
```
